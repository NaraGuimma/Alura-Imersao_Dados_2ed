**Desafio 4**

```
#DESAFIO-4: Plotar os histogramas das idades dos treineiros e nao treineiros - um ao lado do outro
bins = 80
plt.figure(figsize=(10,8))

list1 = dados['NU_IDADE']
plt.hist(list1, bins, color="blue", label="Todos os candidatos")


list2=dados.query('IN_TREINEIRO == 1')['NU_IDADE']
plt.hist(list2, bins, color="yellow", label="Somente treineiros")



plt.show();
```

```
#DESAFIO-4: Plotar os histogramas das idades dos treineiros e nao treineiros - um ao lado do outro - part 2
candidatos=dados.query('IN_TREINEIRO == 0')
treineiro=dados.query('IN_TREINEIRO == 1')
plt.figure(figsize=(12, 6))
sns.kdeplot(x=treineiro['NU_IDADE'], color='blue')
sns.kdeplot(x=candidatos['NU_IDADE'], color='orange')
plt.title('KdePlot das idades entre candidatos e treineiros')
plt.legend(labels=['Treineiros', 'Nao Treineiros'])
plt.show()
```
