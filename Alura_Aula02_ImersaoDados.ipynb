**Desafio 4:** Criar uma função para plotar o boxplot do seaborn

```
def func_boxplot(colunaA,colunaB):
  plt.figure(figsize=(10,8))
  sns.boxplot(x= colunaA, y= colunaB, data=dados, order=renda_ordenada)
  plt.suptitle("Notas de matemática pela renda");
```

```
# removendo as pessoas que nao fizeram nenhuma das provas
dados_sem_notas_zero = dados.query('(NU_NOTA_CN > 0) or (NU_NOTA_CH > 0) or (NU_NOTA_LC > 0) or (NU_NOTA_MT > 0) or (NU_NOTA_REDACAO > 0)')
dados_sem_notas_zero
```

```
def new_boxplot(colA, colB, db):
  plt.figure(figsize = (10,8))
  sns.boxplot(x=colA, y=colB, data=db, order=renda_ordenada)
  plt.suptitle("Somatório das notas pela renda");
```

```
new_boxplot('Q006', 'SOMA_NOTAS', dados_sem_notas_zero)
```

```
new_boxplot('Q006', 'SOMA_NOTAS', dados_sem_notas_zero_2)
```

```
# acrescentar mais dados no boxplot (renda + treineiro/não treineiro)
plt.figure(figsize=(10,6))
sns.boxplot(x= 'Q006', y='SOMA_NOTAS', data=dados_sem_notas_zero, hue='IN_TREINEIRO', order=renda_ordenada)
plt.title('Boxplot da renda pelas notas totais e tipo de candidato - treineiro ou não');
```
